```python
from matplotlib import pyplot as plt
import numpy as np
import scipy.sparse as sparse
from scipy.sparse.csgraph import reverse_cuthill_mckee
```

```python
data = np.load("small-fixture.npy")
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(data[:, 0], data[:, 1], data[:, 2])
```

```python
mat = sparse.coo_matrix((data[:, 2], (data[:, 0], data[:, 1])), shape=(1000, 1000)).tocsr()
```

```python
nm = reverse_cuthill_mckee(mat)
```

```python
mat = (mat[nm, :][:, nm]).tocoo()
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(mat.row, mat.col, mat.data)
```
